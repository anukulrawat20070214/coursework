```python
# problem 1 
def gcdd(a,b):
    while b!=0:
        r=a%b
        a=b
        b=r
    return f" question 1 part 1 ans - the greatest common divisor is {a}" 
gcdd(12,24)
    
 
```

```python
# part 2 ( when no is negative)
def gcd(a,b):
    a=abs(a)
    b=abs(b)
    while b!=0:
        r=a%b
        a=b
        b=r
    return f" question 1 part 2 ans - the greatest common divisor is {a}" 
gcd(24,48) 
```

```python
# part 3
import math
def euler_sum(k):
    sum=0
    for i in range(k+1):
        sum = sum+1/math.factorial(i)
    return sum 
euler_sum(1000)   
```

```python
#part 4
import math 
def euler_approx(a):
    k=0
    total=0
    no=1
    while no>a:
        no=1/math.factorial(k)
        total=total+no
        k+=1
    return total
euler_approx(0.000001)
```

```python
# part 5 
# error functions
# error for sum
def print_euler_sum_error(n):
    approx=euler_sum(n)
    error=approx-2.718
    print("approximation:",approx)
    print("error:",error)
print_euler_sum_error(10)


```

```python
#error for limit
import math
def euler_limit(n):
    return (1+1/n)**n

    
def print_euler_limit_error(n):
    approx=euler_limit(n)
    error=approx-2.718
    print("approximation:",approx)
    print("error:",error)
print_euler_limit_error(100)
```

```python
# question 3 
# part 1 
def is_power2(a):
    while a>1:
        if a%2!=0:
            return False
        a=a//2
    return True
is_power2(10)
```

```python
# part 2
def is_power(b,n):
    while n>1:
        if n%b!=0:
            return False
        n=n//b
    return True
is_power(2,4)
```
